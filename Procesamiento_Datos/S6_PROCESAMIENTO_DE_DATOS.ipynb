Ejemplo 02: Métodos avanzados de indexación

```python
import pandas as pd

data = pd.Series([10, 20, 30, 40, 50], index=['a', 'b', 'c', 'd', 'e'])
```

```python
data.iloc[0]  # Accede al primer elemento
```

```python
data.loc['a']  # Accede al elemento con índice 'a'
```

```python
data.iloc[1:3]  # Elementos en posiciones 1 y 2
```

```python
data.loc[['a', 'b', 'c']]  # Elementos con índices 'a', 'b', y 'c'
```

```python
inventario = pd.Series([300, 200, 150], index=['tornillos', 'tuercas', 'clavos'])

# Selecciona inventario de tornillos y clavos
seleccion = inventario.loc[['tornillos', 'clavos']]

print(seleccion)
# Salida:
# tornillos    300
# clavos       150
# dtype: int64
```

```python
mediciones = pd.Series([0.330, 0.335, 0.329], index=[1, 2, 3])

# Acceder a la segunda medición usando .iloc
segunda_medicion = mediciones.iloc[1]

print(segunda_medicion)
# Salida: 0.335
```

```python
temperaturas = pd.Series([22.5, 23.1, 22.8], index=['9AM', '12PM', '3PM'])

# Acceder a la temperatura a mediodía
temp_mediodia = temperaturas.loc['12PM']

print(temp_mediodia)
# Salida: 23.1
```

```python
niveles_glucosa = pd.Series([90, 140, 130], index=['Paciente 1', 'Paciente 2', 'Paciente 3'])

# Obtener niveles de glucosa de Paciente 2
glucosa_p2 = niveles_glucosa.loc['Paciente 2']

print(glucosa_p2)
# Salida: 140
```

```python
ventas = pd.Series([1000, 1500, 1200], index=['Enero', 'Febrero', 'Marzo'])

# Seleccionar ventas de Enero y Marzo
ventas_em = ventas.loc[['Enero', 'Marzo']]

print(ventas_em)
# Salida:
# Enero    1000
# Marzo    1200
# dtype: int64
```

Ejemplo 03: Dataframes



```python
data = {
    'nombre': ['Juan', 'Ana', 'Luis', 'Diana'],
    'edad': [25, 30, 35, 40],
    'ciudad': ['CDMX', 'GDL', 'MTY', 'PUE']
}
df = pd.DataFrame(data)

print(df)

# Salida:
#    nombre  edad ciudad
# 0   Juan    25   CDMX
# 1    Ana    30    GDL
# 2   Luis    35    MTY
# 3  Diana    40    PUE
```

```python
s1 = pd.Series([1, 2, 3], index=['a', 'b', 'c'], name='Columna1')
s2 = pd.Series([4, 5, 6], index=['a', 'b', 'c'], name='Columna2')
df = pd.DataFrame({s1.name: s1, s2.name: s2})

print(df)

# Salida:
#    Columna1  Columna2
# a         1         4
# b         2         5
# c         3         6
```

```python
data = {
    'nombre': ['Juan', 'Ana', 'Luis', 'Diana'],
    'edad': [25, 30, 35, 40],
    'ciudad': ['CDMX', 'GDL', 'MTY', 'PUE']
}
df = pd.DataFrame(data, index=['renglon_1', 'renglon_2', 'renglon_3', 'renglon_4'])

print(df)

# Salida:
#             nombre  edad  ciudad
# renglon_1    Juan    25    CDMX
# renglon_2    Ana     30    GDL
# renglon_3    Luis    35    MTY
# renglon_4    Diana   40    PUE
```

```python
print(df[['nombre', 'ciudad']])

# Salida:
#    nombre ciudad
# 0   Juan   CDMX
# 1    Ana    GDL
# 2   Luis    MTY
# 3  Diana    PUE
```

```python
 print(df.iloc[1:3])

 # Salida:
 #    nombre  edad ciudad
 # 1    Ana    30    GDL
 # 2   Luis    35    MTY

 # Recuerda que el rango es inclusivo en el inicio y exclusivo en el final.
```

```python
 print(df.loc[1:3, ['nombre', 'ciudad']])

 # Salida:
 #    nombre ciudad
 # 1    Ana    GDL
 # 2   Luis    MTY
 # 3  Diana    PUE

 # En este caso, el primer rango corresponde a las filas y el segundo rango a las columnas.
```

Ejemplo 04: Manipulación de columnas

```python
import pandas as pd

# Diccionario datos de ejemplo
datos = {
   "ID": [101, 102, 103, 104, 105],
   "Nombre": ["Ana", "Luis", "Marta", "Juan", "Sofía"],
   "Departamento": ["Finanzas", "Marketing", "TI", "Recursos Humanos", "Operaciones"],
   "Salario": [45000, 38000, 62000, 41000, 36000]
   }

# Crear DataFrame
df = pd.DataFrame(datos)

# Añadir columna "Edad"
df["Edad"] = [25, 30, 28, 35, 32]
print(df)
```

```python

# Reasigna nuevos valores a la columna "Salario"
df["Salario"] = [47000, 40000, 64000, 43000, 37000]
print(df)
```

```python
# Añadir columna "Tiempo"
df["Tiempo"] = [2, 3, 4, 5, 6]
print(df)
```

```python
df = df.drop(columns=["Tiempo"])
print(df)
```

```python
df = df.drop("Tiempo", axis=1, errors="ignore")
print(df)
```

```python
# Diccionario de gastos
datos = {
    "Marketing": 32000,
    "Ventas": 45000,
    "Desarrollo de Producto": 62000,
    "Soporte Técnico": 25000,
    "Recursos Humanos": 15000,
    "IT": 50000,
    "Operaciones": 47000,
    "Investigación y Desarrollo": 55000
}
# Crear la serie
gastos_mensuales = pd.Series(datos)
print("Tamaño:", gastos_mensuales.size)
print("Tipo de dato:", gastos_mensuales.dtype)
print("Divisiones incluidas:")
print(gastos_mensuales.index)
print("\nGastos de Ventas, IT e Investigación y Desarrollo:")
print(gastos_mensuales[["Ventas", "IT", "Investigación y Desarrollo"]])
print("\nGastos desde Marketing hasta Recursos Humanos:")
print(gastos_mensuales["Marketing":"Recursos Humanos"])
print("\nGastos desde Desarrollo de Producto hasta Operaciones:")
print(gastos_mensuales["Desarrollo de Producto":"Operaciones"])
print("\nGastos desde Soporte Técnico hasta el final:")
print(gastos_mensuales["Soporte Técnico":])
#Extra
print(f"\nGasto total de la empresa: ${gastos_mensuales.sum()}")
print(f"Gasto promedio por división: ${gastos_mensuales.mean()}")
mayor_gasto = gastos_mensuales.idxmax()
print(f"División con mayor gasto: {mayor_gasto} (${gastos_mensuales[mayor_gasto]})")
menor_gasto = gastos_mensuales.idxmin()
print(f"División con menor gasto: {menor_gasto} (${gastos_mensuales[menor_gasto]})")
```
