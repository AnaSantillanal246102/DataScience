Ejemplo 01: Manipulaci贸n de tipos de datos y strings

```python
import pandas as pd

# Crear un DataFrame de ejemplo
df = pd.DataFrame({
    'A': ['1', '2', '3', '4'],
    'B': [1.1, 2.2, 3.3, 4.4],
    'C': ['2021-01-01', '2021-02-01', '2021-03-01', '2021-04-01'],
    'D': [1212883200000, 1296518400000, 1327881600000, 1361923200000]
})

# Convertir la columna 'A' a tipo entero
df['A'] = df['A'].astype(int)

# Convertir la columna 'C' a tipo datetime usando pd.to_datetime
df['C'] = pd.to_datetime(df['C'])

# Convertir la columna 'D' a tipo datetime en milisegundos
df = df.astype({'D': 'datetime64[ms]'})

# Mostrar los tipos de datos para verificar las conversiones
df.dtypes

# Salida esperada:
# A             int64
# B           float64
# C    datetime64[ns]
# D    datetime64[ns]
# dtype: object
```

```python
# Ordenar renglones por las columnas 'Edad' y 'D铆as Hospitalizado'
df_ordenado = df.sort_values(by=['Edad', 'D铆as Hospitalizado'], ascending=[True, False])
df_ordenado.head(10)
```

Transformaci贸n con map y apply

```python
# Supongamos que necesitamos codificar la columna de 'Diagn贸stico' para un an谩lisis an贸nimo.
diagnostico_codificado = {
    'Apendicitis': 'D01',
    'Fractura de brazo': 'D02',
    'Gripe': 'D03',
    'Diabetes Tipo 2': 'D04',
    'Hipertensi贸n': 'D05',
    'Alergias': 'D06',
    'Anemia': 'D07',
    'Bronquitis': 'D08',
    'Artritis': 'D09',
    'Quemaduras leves': 'D10'
}
df['Diagn贸stico Codificado'] = df['Diagn贸stico'].map(diagnostico_codificado)
df[['Diagn贸stico', 'Diagn贸stico Codificado']]
```

```python
# Funci贸n para categorizar edades
def categorizar_edad(edad):
    if edad < 30:
        return '20-29 a帽os'
    elif edad < 40:
        return '30-39 a帽os'
    elif edad < 50:
        return '40-49 a帽os'
    else:
        return '50+ a帽os'

# Aplicar map con una funci贸n para transformar las edades
df['Grupo de Edad'] = df['Edad'].map(categorizar_edad)
df[['Paciente', 'Edad', 'Grupo de Edad']]
```

Uso de apply en un dataframe:


```python
# Calcular el costo estimado del tratamiento por d铆a hospitalizado
df['Costo por D铆a'] = df.apply(lambda row: round(row['Costo del Tratamiento ($)'] / row['D铆as Hospitalizado'], 2), axis=1)
df[['Paciente', 'Costo del Tratamiento ($)', 'D铆as Hospitalizado', 'Costo por D铆a']]
```

```python
# Ajustar el costo del tratamiento basado en el n煤mero de d铆as hospitalizados
def ajustar_costo(row):
    descuento = 100 * (row['D铆as Hospitalizado'] - 1)  # $100 descuento por cada d铆a 'ADICIONAL', no por el primero.
    return row['Costo del Tratamiento ($)'] - descuento

# Aplicar la funci贸n a lo largo de los renglones
df['Costo Ajustado ($)'] = df.apply(ajustar_costo, axis=1)
df[['Paciente', 'D铆as Hospitalizado', 'Costo del Tratamiento ($)', 'Costo Ajustado ($)']].head(10)
```

# Ejemplo 04: Combinaci贸n y agrupaci贸n de datos

```python
import pandas as pd

# DataFrame de Clientes
clientes = pd.DataFrame({
    'cliente_id': [101, 102, 103, 104],
    'nombre': ['Alice Johnson', 'Bob Smith', 'Charlie Davis', 'Diana Prince'],
    'email': ['alice@example.com', 'bob@example.com', 'charlie@example.com', 'diana@example.com'],
    'regi贸n': ['Norte', 'Sur', 'Este', 'Oeste']
})
clientes.head()
```

```python
# DataFrame de Pedidos
pedidos = pd.DataFrame({
    'pedido_id': ['P001', 'P002', 'P003', 'P004', 'P005', 'P006'],
    'cliente_id': [101, 103, 105, 104, 101, 102],  # Incluye IDs que no existen y repeticiones
    'total': [250, 150, 450, 300, 320, 110],
    'fecha': ['2021-01-10', '2021-02-15', '2021-03-20', '2021-04-22', '2021-05-25', '2021-06-15'],
    'categor铆a': ['Electr贸nica', 'Ropa', 'Electr贸nica', 'Libros', 'Alimentos', 'Ropa']
})
pedidos.head(10)
```

```python
# Inner Join
df_inner = pd.merge(clientes, pedidos, on='cliente_id', how='inner')
df_inner.head(10)
```

```python
# Left Join
df_left = pd.merge(clientes, pedidos, on='cliente_id', how='left')
df_left.head(10)
```

```python
# Right Join
df_right = pd.merge(clientes, pedidos, on='cliente_id', how='right')
df_right.head(10)
```

group by

```python
# Realizamos un left join para obtener un DataFrame completo
df_merged = pd.merge(clientes, pedidos, on='cliente_id', how='left')
df_merged.head(10)
```

```python
# Ejemplo 1: Total de Pedidos por Regi贸n
total_por_regi贸n = df_merged.groupby('regi贸n')['total'].sum()
print(total_por_regi贸n)

# Ejemplo 2: N煤mero de Pedidos por Categor铆a y Regi贸n
conteo_pedidos_categor铆a = df_merged.groupby(['regi贸n', 'categor铆a'])['pedido_id'].count()
print(conteo_pedidos_categor铆a)

# Ejemplo 3: Promedio del Total de Pedidos por Mes
# Convertir 'fecha' a datetime
df_merged['fecha'] = pd.to_datetime(df_merged['fecha'])

# Extraer el mes y a帽o de la fecha
df_merged['mes_a帽o'] = df_merged['fecha'].dt.to_period('M')

# Calcular el promedio de ventas por mes
promedio_ventas_mes = df_merged.groupby('mes_a帽o')['total'].mean()
print(promedio_ventas_mes)
```

```python
import pandas as pd
import numpy as np

# Generar datos para el DataFrame de tr谩fico
np.random.seed(0)
fechas_trafico = pd.date_range(start='2024-07-01', periods=30, freq='D')
horas_trafico = ['08:00', '17:00']
zonas_trafico = ['Centro', 'Norte', 'Sur', 'Este', 'Oeste']

# Crear DataFrame de tr谩fico con 150 datos
df_trafico = pd.DataFrame(
    {
      'fecha': np.tile(fechas_trafico, len(horas_trafico) * len(zonas_trafico)),
      'hora': np.repeat(horas_trafico * len(fechas_trafico), len(zonas_trafico)),
      'zona': np.tile(zonas_trafico, len(fechas_trafico) * len(horas_trafico)),
      'vehiculos': np.random.randint(100, 500, size=(len(fechas_trafico) * len(horas_trafico) * len(zonas_trafico)))
  }
)

# Crear DataFrame meteorol贸gico con 30 d铆as de datos
df_meteorologico = pd.DataFrame(
    {
      'fecha': fechas_trafico,
      'temperatura': np.random.uniform(25, 35, size=len(fechas_trafico)),
      'precipitacion': np.random.uniform(0, 10, size=len(fechas_trafico))
  }
)
df_trafico.sort_values(['zona', 'hora'])
df_trafico['zona'] = df_trafico['zona'].str.strip()
df_trafico['vehiculos'] = df_trafico['vehiculos'].apply(lambda x: x / 1000)
df_trafico.head()
```

```python
df_combinado=pd.merge(df_trafico, df_meteorologico, on='fecha')
df_combinado.head()
```

```python
df_agrupado = df_combinado.groupby(['zona', 'hora']).agg({
    'vehiculos': 'mean',
    'temperatura': 'max'
}).reset_index()
df_agrupado.head()
```
